```python
import tensorflow as tf
import numpy as np

tf.__version__
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from keras.models import load_model
from keras.layers import LSTM
from keras.utils.vis_utils import plot_model
from keras.optimizers import Adam
```

```python
npz=np.load(r'C:\Users\Hrushikesh\OneDrive\Desktop\Data Science course\Projects\ML Project\Data DL NPZ\Stock Prediction Train.npz')
train_inputs=npz['inputs']
train_targets=npz['targets']

npz=np.load(r'C:\Users\Hrushikesh\OneDrive\Desktop\Data Science course\Projects\ML Project\Data DL NPZ\Stock Prediction Validate.npz')
validation_inputs, validation_targets=npz['inputs'], npz['targets']

npz=np.load(r'C:\Users\Hrushikesh\OneDrive\Desktop\Data Science course\Projects\ML Project\Data DL NPZ\Stock Prediction Test.npz')
test_inputs, test_targets=npz['inputs'], npz['targets']

```

```python
X_train = np.reshape(train_inputs, (train_inputs.shape[0], train_inputs.shape[1], 1))
X_val = np.reshape(validation_inputs, (validation_inputs.shape[0], validation_inputs.shape[1], 1))
X_test = np.reshape(test_inputs, (test_inputs.shape[0], test_inputs.shape[1], 1))
```

```python
from keras.layers import Dropout
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))  # Dropout layer for regularization
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=25))  # Dense layer with 25 neurons
model.add(Dense(units=1))   # Final output layer for prediction
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X_train, train_targets, 
                    epochs=20, 
                    batch_size=32, 
                    validation_data=(X_val, validation_targets), 
                    verbose=1)
```

```python
test_loss = model.evaluate(X_test, test_targets, verbose=1)
test_loss
```

```python
y_pred = model.predict(X_test)
```
